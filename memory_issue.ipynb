```python
import torch

# 清空CUDA缓存
torch.cuda.empty_cache()

# 创建一个大Tensor
x = torch.randn(10000, 10000, device='cuda')  
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

# 创建中间变量
y = x * 2
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

# 再次清空CUDA缓存
torch.cuda.empty_cache()
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

# 第二次测试，清空CUDA缓存
torch.cuda.empty_cache()

# 创建一个大Tensor
z = torch.randn(10000, 10000, device='cuda')  
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

# 创建中间变量
u = z * 2
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

# 再次清空CUDA缓存
torch.cuda.empty_cache()
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

del u,z

print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存
try: 
    z = z**2
    print('z not_deleted') 
except: print('z deleted')

print('By del tensors, we release memories')

x=x.to('cpu');y=y.to('cpu')
try: 
    x = x**2
    print('x not_deleted') 
except: print('x deleted')

print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存

def function_with_intermediate():
    x = torch.randn(10000, 10000, device='cuda')  # 创建一个更大的Tensor
    print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存
    x = x**2
    print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存
    y = x*2+x**2  # 中间变量
    print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存
    return y

y = function_with_intermediate()
print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024 ** 2:.2f} MB")  # 打印当前分配的显存
```
